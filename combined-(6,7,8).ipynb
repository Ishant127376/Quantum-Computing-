```python
# ==============================
# LAB 8 – EXPERIMENT 1
# Phase Kickback
# ==============================

# Install qiskit if not already installed
!pip install qiskit

from qiskit import QuantumCircuit

print("===== LAB 8 EXPERIMENT 1: PHASE KICKBACK =====")

# Reverse CNOT using phase kickback idea
qc = QuantumCircuit(2)

qc.h(0)
qc.h(1)

qc.cx(1, 0)   # reverse control

qc.h(0)
qc.h(1)

print("Reverse CNOT Circuit:")
print(qc.draw())


# Kickback using T gate
qc2 = QuantumCircuit(2)

qc2.h(0)
qc2.t(1)

qc2.cx(0, 1)

qc2.h(0)

print("\nPhase Kickback using T Gate:")
print(qc2.draw())
```

```python
# ==============================
# LAB 8 – EXPERIMENT 2
# Deutsch Algorithm
# ==============================

!pip install qiskit-aer

from qiskit import QuantumCircuit
from qiskit_aer import Aer

print("===== LAB 8 EXPERIMENT 2: DEUTSCH ALGORITHM ====")

qc = QuantumCircuit(2,1)

# Initialize
qc.x(1)
qc.h(0)
qc.h(1)

# Oracle (balanced example)
qc.cx(0,1)

# Measurement
qc.h(0)
qc.measure(0,0)

print(qc.draw())

sim = Aer.get_backend('aer_simulator')
result = sim.run(qc).result()

print("Output:", result.get_counts())

print("\nInterpretation:")
print("0 → Constant function")
print("1 → Balanced function")
```

```python
# ==============================
# LAB 8 – EXPERIMENT 3
# Deutsch-Jozsa Algorithm
# ==============================

from qiskit import QuantumCircuit
from qiskit_aer import Aer

print("===== LAB 8 EXPERIMENT 3: DEUTSCH-JOZSA =====")

n = 3   # number of input qubits

qc = QuantumCircuit(n+1, n)

# Step 1: Initialize
qc.x(n)
qc.h(range(n+1))

# Step 2: Oracle (balanced example)
for i in range(n):
    qc.cx(i, n)

# Step 3: Apply H again
qc.h(range(n))

# Step 4: Measure
qc.measure(range(n), range(n))

print(qc.draw())

# Run
sim = Aer.get_backend('aer_simulator')
result = sim.run(qc).result()

print("Output:", result.get_counts())

print("\nInterpretation:")
print("All zeros → Constant")
print("Anything else → Balanced")

```

```python
# ==============================
# LAB 7 – EXPERIMENT 1
# Superdense Coding
# ==============================

from qiskit import QuantumCircuit
from qiskit_aer import Aer

print("===== LAB 7 EXPERIMENT 1: SUPERDENSE CODING =====")

qc = QuantumCircuit(2,2)

# Step 1: Create Bell pair
qc.h(0)
qc.cx(0,1)

# Step 2: Encode message (example: '10')
qc.z(0)

# Step 3: Decode
qc.cx(0,1)
qc.h(0)

# Step 4: Measure
qc.measure([0,1],[0,1])

print(qc.draw())

sim = Aer.get_backend('aer_simulator')
result = sim.run(qc).result()

print("Output:", result.get_counts())

```

```python
# ==============================
# LAB 7 – EXPERIMENT 2
# Quantum Teleportation
# ==============================

from qiskit import QuantumCircuit
from qiskit_aer import Aer

print("=====LAB 7 EXPERIMENT 2: QUANTUM TELEPORTATION =====")

qc = QuantumCircuit(3,3)

# Step 1: Create state to teleport
qc.h(0)

# Step 2: Create Bell pair
qc.h(1)
qc.cx(1,2)

# Step 3: Bell measurement
qc.cx(0,1)
qc.h(0)

qc.measure([0,1],[0,1])

# Step 4: Apply corrections
qc.cx(1,2)
qc.cz(0,2)

qc.measure(2,2)

print(qc.draw())

sim = Aer.get_backend('aer_simulator')
result = sim.run(qc).result()

print("Output:", result.get_counts())

```

```python
# ==============================
# LAB 7 – EXPERIMENT 3
# Quantum Coin Game
# ==============================

from qiskit import QuantumCircuit
from qiskit_aer import Aer

print("=====LAB 7 EXPERIMENT 3: QUANTUM COIN GAME =====")

qc = QuantumCircuit(1,1)

# Quantum strategy
qc.h(0)    # superposition
qc.z(0)    # phase flip
qc.h(0)

qc.measure(0,0)

print(qc.draw())

sim = Aer.get_backend('aer_simulator')
result = sim.run(qc).result()

print("Output:", result.get_counts())

```

```python
# ==============================
# LAB 6 – EXPERIMENT 1
# X-Gate on Bloch Sphere and State Vector
# ==============================

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
import matplotlib.pyplot as plt

print("=====LAB 6 EXPERIMENT 1: X GATE =====")

qc = QuantumCircuit(1)
qc.x(0)

print("Circuit:")
print(qc.draw())

state = Statevector.from_instruction(qc)
print("Statevector:", state)

plot_bloch_multivector(state)
plt.show()

```

```python
# ==============================
# LAB 6 – EXPERIMENT 2
# Single Qubit Gates on Bloch Sphere and State Vector
# ==============================

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
import matplotlib.pyplot as plt

print("===== ;AB 6 EXPERIMENT 2: Y, Z, PHASE, S, Sdg, T, Tdg =====")

gates = ['y', 'z', 'p', 's', 'sdg', 't', 'tdg']

for gate in gates:
    qc = QuantumCircuit(1)

    # Initial state = 1/sqrt(2) (|0> + |1>) = |+>
    qc.h(0)

    if gate == 'p':
        qc.p(3.14159/4, 0)   # example phase angle
        gate_name = "PHASE"
    else:
        getattr(qc, gate)(0)
        gate_name = gate.upper()

    print(f"\n===== Gate: {gate_name} =====")
    print(qc.draw())

    state = Statevector.from_instruction(qc)
    print("Statevector:", state)

    plot_bloch_multivector(state)
    plt.show()
```

```python
# ==============================
# LAB 6 – EXPERIMENT 3
# Introduction to State Vector Simulator
# ==============================

from qiskit import QuantumCircuit
from qiskit_aer import Aer

print("===== LAB 6 EXPERIMENT 3 =====")

sim = Aer.get_backend('aer_simulator')

# (a) H Z H
qc_a = QuantumCircuit(1, 1)
qc_a.h(0)
qc_a.z(0)
qc_a.h(0)
qc_a.measure(0, 0)

print("\nCircuit (a):")
print(qc_a.draw())
result_a = sim.run(qc_a, shots=1024).result()
print("Output (a):", result_a.get_counts())

# (b) H X H
qc_b = QuantumCircuit(1, 1)
qc_b.h(0)
qc_b.x(0)
qc_b.h(0)
qc_b.measure(0, 0)

print("\nCircuit (b):")
print(qc_b.draw())
result_b = sim.run(qc_b, shots=1024).result()
print("Output (b):", result_b.get_counts())

# (c) Two-qubit entangled circuit
qc_c = QuantumCircuit(2, 2)
qc_c.h(0)
qc_c.cx(0, 1)
qc_c.measure([0, 1], [0, 1])

print("\nCircuit (c):")
print(qc_c.draw())
result_c = sim.run(qc_c, shots=1024).result()
print("Output (c):", result_c.get_counts())

# (d) Another two-qubit circuit
qc_d = QuantumCircuit(2, 2)
qc_d.x(0)
qc_d.cx(0, 1)
qc_d.measure([0, 1], [0, 1])

print("\nCircuit (d):")
print(qc_d.draw())
result_d = sim.run(qc_d, shots=1024).result()
print("Output (d):", result_d.get_counts())
```
